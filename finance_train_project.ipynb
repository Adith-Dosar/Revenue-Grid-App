# Data Dictionary

# Label Encoding

```python
#Gender

data_train['gender'].replace(to_replace=['Male','Female','Unknown'],value=[1,2,0],inplace=True)

#Children

data_train['children'].replace(to_replace=['Zero','1','2','3','4+'],value=[0,1,2,3,4],inplace=True)

#Importing Label Encoder

from sklearn import preprocessing
label_encoder=preprocessing.LabelEncoder()

#Family Income

data_train['family_income']=label_encoder.fit_transform(data_train['family_income'])

#Age Band

data_train['age_band']=label_encoder.fit_transform(data_train['age_band'])

#Status

data_train['status']=label_encoder.fit_transform(data_train['status'])

#Occupation

data_train['occupation']=label_encoder.fit_transform(data_train['occupation'])

#Loop for Label Encoding

q=[]
for i in data.columns:
    if data[i].dtypes=='object':
        q.append(i)
#print(q)
#len(q)

q.remove('post_area')
q.remove('post_code')
q.remove('age_band')
q.remove('family_income')
q.remove('children')
len(q)

for i in q:
    data_train[i]=label_encoder.fit_transform(data_train[i])
    #print(i)
```

# Droping Irrelevant Variables

```python
#Deep Copy of Data
copy_data=data_train.copy()
```

```python
copy_data.drop(['REF_NO','post_area','post_code'],axis=1,inplace=True)

copy_data.drop('year_last_moved',axis=1,inplace=True)
```

# Splitting Data in Train and Test

```python
from sklearn.model_selection import train_test_split
train,test=train_test_split(copy_data,train_size=0.8,random_state=123)
```

```python
train['source']='train'
test['source']='test'
```

```python
trainX=train.loc[:,train.columns!='Revenue.Grid']
trainY=train.loc[:,'Revenue.Grid']
testX=test.loc[:,test.columns!='Revenue.Grid']
testY=test.loc[:,'Revenue.Grid']
```

```python
trainX.drop('source',axis=1)
testX.drop('source',axis=1)
```

# Correlation Heatmap

```python
#Correlation
plt.subplots(figsize=(22,22))
sns.heatmap(copy_data.corr(),annot=True,cmap='inferno_r')
```

```python
#Deep Copy of trainX and testX
x_train=trainX.copy()
x_test=testX.copy()
```

```python
x_train.drop('source',inplace=True,axis=1)
x_test.drop('source',inplace=True,axis=1)
```

# Renaming the Features

```python
a=[i.upper().replace('.','_') for i in x_train.columns]
x_train.rename(columns=dict(zip(x_train.columns,a)),inplace=True)
x_test.rename(columns=dict(zip(x_test.columns,a)),inplace=True)
```

# Model Building (Random Forest)

```python
#Importing Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=1000,random_state=123,n_jobs=-1)
```

```python
#Model Training
rf.fit(x_train,trainY)
```

```python
#Model Prediction
rf_pred=rf.predict(x_test)
```

```python
#Model Accuracy
from sklearn.metrics import accuracy_score,confusion_matrix, classification_report
#rf_accuracy=accuracy_score(testY,rf_pred)
#print('Accuracy:',rf_accuracy)
print('Classification Report :')
print()
print(classification_report(testY,rf_pred))
print()
Confusion_Mat = confusion_matrix(testY, rf_pred)
print('Confusion Matrix :')
print()
print(Confusion_Mat)
```

# Model Building (KNN)

```python
#Importing KNearest Neighbors Classsifier
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=10,metric='minkowski',leaf_size=30,algorithm='auto',p=2)
```

```python
#Model Training
knn.fit(x_train,trainY)
```

```python
#Model Prediction
knn_pred=knn.predict(x_test)
```

```python
#Model Accuracy
from sklearn.metrics import accuracy_score,confusion_matrix, classification_report
#knn_accuracy=accuracy_score(testY,knn_pred)
#print('Accuracy:',knn_accuracy)
print('Classification Report :')
print()
print(classification_report(testY,knn_pred))
print()
Confusion_Mat = confusion_matrix(testY, knn_pred)
print('Confusion Matrix :')
print()
print(Confusion_Mat)
```

# Grid Search Cross Validation

```python
from sklearn.model_selection import GridSearchCV

NN = range(1,14,2)
P = range(1,4,1) #This "p" is "h" in minkowski formula
param_grid = {'n_neighbors': NN, 'p': P} # param_grid is a dictionary 

Grid_Search_Model = GridSearchCV(estimator = KNeighborsClassifier(), 
                     param_grid=param_grid,  
                     scoring='accuracy', 
                     cv=5,n_jobs=-1).fit(x_train, trainY)
```

```python
df=pd.DataFrame.from_dict(Grid_Search_Model.cv_results_)
df.to_csv("GridSearchKNN.csv")
```
